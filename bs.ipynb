```python
import requests
from bs4 import BeautifulSoup
import os

from webdriver_manager.chrome import ChromeDriverManager
import requests
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

import pandas as pd


#options
options = Options()
options.add_experimental_option("detach", True)

#initialise webdriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), 
                          options=options)

# Enter the Google search query
query = "Data Analyst"

# Format the query for use in a URL
url_query = query.replace(" ", "+")

# Set the URL for the Google search
url = f"https://www.google.com/search?q={url_query}&ibp=htl;jobs"
#url = f"file:///C:/Users/Ben/Documents/python/DancinParrotAutoForm/AutoForm/testing.html"
# Set the headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"}
# Send a GET request to the URL with the headers
driver.get(url)
time.sleep(5) 

# soup = BeautifulSoup(driver.page_source, "html.parser")
# print(soup.text[:20])
# l = soup.find('ul')

# job_listings = soup.find_all("li")

details = []
xpath = "//div[@class='PwjeAc']"
listings = driver.find_elements(By.XPATH, xpath)


def click(job):
    
    time.sleep(1)
    try:
        job.click()
        
    except Exception as e:
        print( f'click err: {e}')
        driver.maximize_window()
        driver.execute_script("arguments[0].scrollIntoView(true);", job)
        click(job)
```

```python
        
# Print the job titles and URLs
#for job in job_listings:
elem = {'title':[],
        'company':[],
        'location':[],
        'link':[],
        'type':[],
        'resp':[]
        }
for i, job in enumerate(listings):
    ##Elements from side list
    elem['company'].append(job.find_element(By.CLASS_NAME, 'vNEEBe').text)
    elem['location'].append(job.find_element(By.CLASS_NAME, 'Qk80Jf').text)
    elem['title'].append(job.find_element(By.CLASS_NAME, 'Fol1qc').text)
    time.sleep(1)
    try:
        click(job)

        print('clicked')
        time.sleep(1)
        
    except Exception as e:
        print(f'click er: {e}')
    try:
        links=[]
        applies = driver.find_elements(By.XPATH,"//a[@href]")
        for link in applies:
            l = link.get_attribute("href")
            #print(l)
            if 'utm_campaign=google_jobs_apply' in l:
                links.append(l)
        elem['link'].append(links)
        elem['type'].append(driver.find_element(By.CLASS_NAME, 'KKh3md').text)
        elem['resp'].append(driver.find_element(By.CLASS_NAME, 'JxVj3d').text)
    except Exception as e:
        print(f'desc er: {e}')
    print(elem)
    print()
    time.sleep(5)
    
```
